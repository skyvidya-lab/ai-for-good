One-time setup for the Dynamis Terra Colab environment.

```python
# 1. Mount Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=False)
WORKSPACE = '/content/drive/Shareddrives/SKYVIDYA/AI for Good/datasets_final_round'
print('WORKSPACE:', WORKSPACE)
```

```python
# 2. Verify all zips
from pathlib import Path
EXPECTED = [
    ('track1_download_link_1.zip', 129 * 1024 * 1024),         # 129 MB
    ('track1_download_link_2.zip', 17.48 * 1024**3),
    ('track1_download_link_3.zip', 17.49 * 1024**3),
    ('track1_download_link_4.zip', 17.48 * 1024**3),
    ('track1_download_link_5.zip', 17.48 * 1024**3),
]
for name, expected_size in EXPECTED:
    p = Path(WORKSPACE) / name
    if p.exists():
        size_gb = p.stat().st_size / (1024**3)
        print(f'  ✓ {name}: {size_gb:.2f} GB')
    else:
        print(f'  ✗ {name}: MISSING')
```

```python
# 3. Clone repo (fallback: gdown file IDs if Drive mount fails)
import os
REPO_PATH = '/content/ai-for-good'
REPO_URL = 'YOUR_REPO_URL_HERE'  # TODO: set when pushed to GitHub

if not Path(REPO_PATH).exists():
    if REPO_URL.startswith('YOUR'):
        print('[ACTION] Set REPO_URL above or upload src/ manually')
    else:
        os.system(f'git clone {REPO_URL} {REPO_PATH}')
else:
    os.system(f'cd {REPO_PATH} && git pull --ff-only')

import sys; sys.path.insert(0, REPO_PATH)
print('sys.path contains repo:', REPO_PATH in sys.path)
```

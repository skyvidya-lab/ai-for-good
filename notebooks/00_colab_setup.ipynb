# 00 — Colab Setup

One-time setup for the Dynamis Terra Colab environment.

**Steps:**
1. Mount Google Drive (SKYVIDYA Shared Drive).
2. Verify all 5 download zips are present.
3. Clone / update the code repo.
4. Install Python dependencies.
5. Verify GPU.

Run this before `02_baseline_vs_dynamis.ipynb` or `03_full_training.ipynb`.

```python
# 1. Mount Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=False)
WORKSPACE = '/content/drive/Shareddrives/SKYVIDYA/AI for Good/datasets_final_round'
print('WORKSPACE:', WORKSPACE)
```

```python
# 2. Verify all zips
from pathlib import Path
EXPECTED = [
    ('track1_download_link_1.zip', 129 * 1024 * 1024),         # 129 MB
    ('track1_download_link_2.zip', 17.48 * 1024**3),
    ('track1_download_link_3.zip', 17.49 * 1024**3),
    ('track1_download_link_4.zip', 17.48 * 1024**3),
    ('track1_download_link_5.zip', 17.48 * 1024**3),
]
for name, expected_size in EXPECTED:
    p = Path(WORKSPACE) / name
    if p.exists():
        size_gb = p.stat().st_size / (1024**3)
        print(f'  ✓ {name}: {size_gb:.2f} GB')
    else:
        print(f'  ✗ {name}: MISSING')
```

```python
# 3. Clone (or update) private repo using Colab Secrets
#
# One-time setup:
#   1. Go to GitHub → Settings → Developer settings → Personal access tokens → Fine-grained
#   2. Create a token with "Contents: Read" on GeoProjectAI/ai-for-good
#   3. In Colab, click the key icon on the left sidebar → Add secret:
#        Name: GITHUB_TOKEN
#        Value: <your PAT>
#        Enable "Notebook access"
import os
from google.colab import userdata

REPO_PATH = '/content/ai-for-good'
REPO_OWNER = 'GeoProjectAI'
REPO_NAME = 'ai-for-good'

try:
    token = userdata.get('GITHUB_TOKEN')
except Exception as e:
    token = None
    print(f'[warn] could not read GITHUB_TOKEN secret: {e}')

if token:
    REPO_URL = f'https://x-access-token:{token}@github.com/{REPO_OWNER}/{REPO_NAME}.git'
else:
    REPO_URL = f'https://github.com/{REPO_OWNER}/{REPO_NAME}.git'
    print('[warn] No GITHUB_TOKEN secret — clone will fail for a private repo')

if not Path(REPO_PATH).exists():
    os.system(f'git clone {REPO_URL} {REPO_PATH}')
else:
    os.system(f'cd {REPO_PATH} && git pull --ff-only')

import sys; sys.path.insert(0, REPO_PATH)
print('sys.path contains repo:', REPO_PATH in sys.path)
```
